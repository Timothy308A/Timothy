```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, mean_squared_error, roc_auc_score

# Load the standardized dataset
df = pd.read_csv('Admission_Predict_Ver1.1_standardized.csv')

# Define features and target variable
features = ['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA', 'Research']
X = df[features]
y = df['Chance of Admit ']

# Convert the target variable to binary (0 and 1) if needed
# For example, if you want to classify 'Chance of Admit ' into high/low categories
# Threshold should be defined based on your criteria
threshold = 0.5
y_binary = np.where(y >= threshold, 1, 0)

# Split the data into training and testing sets
trainX, testX, trainY, testY = train_test_split(X, y_binary, train_size=0.7, random_state=5)

# Initialize and fit the Logistic Regression model
log_reg = LogisticRegression()
log_reg.fit(trainX, trainY)

# Predict using the test set
predY = log_reg.predict(testX)
pred_proba = log_reg.predict_proba(testX)[:, 1]  # Probability estimates for ROC AUC

# Calculate evaluation metrics
accuracy = accuracy_score(testY, predY)
roc_auc = roc_auc_score(testY, pred_proba)
mse = mean_squared_error(testY, predY)

# Output the results
print('Accuracy: ', accuracy)
print('ROC AUC Score: ', roc_auc)
print('Mean Squared Error: ', mse)
```

```python
# Extract the coefficients and create a DataFrame for plotting
coefficients = log_reg.coef_[0]
coef_df = pd.DataFrame({'Feature': features, 'Coefficient': coefficients})

# Plot the feature contributions
plt.figure(figsize=(10, 6))
# Use absolute values for plotting
coef_df['Absolute Coefficient'] = coef_df['Coefficient'].abs()
sns.barplot(x="Absolute Coefficient", y="Feature", data=coef_df.sort_values(by="Absolute Coefficient", ascending=False))
plt.title('Feature Importance from Logistic Regression')
plt.xlabel('Absolute Coefficient Value')
plt.ylabel('Feature')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# Load the standardized dataset
df = pd.read_csv('Admission_Predict_Ver1.1.csv')

# Define features and target variable
features = ['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA', 'Research']
X = df[features]
y = df['Chance of Admit ']

# Split the data into training and testing sets
trainX, testX, trainY, testY = train_test_split(X, y, train_size=0.7, random_state=5)

# Initialize and fit the Linear Regression model
lin_reg = LinearRegression()
lin_reg.fit(trainX, trainY)

# Predict using the test set
predY = lin_reg.predict(testX)

# Calculate evaluation metrics
r2 = r2_score(testY, predY)
mse = mean_squared_error(testY, predY)
rmse = np.sqrt(mse)

# Output the results
print('R-Squared (Accuracy): ', r2)
print('Mean Squared Error: ', mse)
print('Root Mean Squared Error: ', rmse)

# Plot Actual vs Predicted Values
plt.figure(figsize=(10, 6))
plt.scatter(testY, predY, color='blue', alpha=0.5, label='Predicted vs Actual')
plt.plot([testY.min(), testY.max()], [testY.min(), testY.max()], color='red', linestyle='--', label='Perfect Fit')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual Values vs Predicted Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Initialize and fit the Linear Regression model
lin_reg = LinearRegression()
lin_reg.fit(trainX, trainY)

# Predict using the test set
predY = lin_reg.predict(testX)

# Calculate residuals
residuals = testY - predY

# Calculate evaluation metrics
r2 = r2_score(testY, predY)
mse = mean_squared_error(testY, predY)
rmse = np.sqrt(mse)

# Output the results
print('R-Squared (Accuracy): ', r2)
print('Mean Squared Error: ', mse)
print('Root Mean Squared Error: ', rmse)

# Plot residuals histogram
plt.figure(figsize=(10, 6))
plt.hist(residuals, bins=30, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.grid(True)
plt.show()
```
